```python
pdf = open("Exercise_Files/Find_the_Phone_Number.pdf", "rb")
```

```python
import PyPDF2
pdf_reader = PyPDF2.PdfFileReader(pdf)
```

```python
pdf_reader.numPages
```

```python
import re
phone_nr = ""
pattern = r'(\d\s?){3} (\d\s?){3}'

for page in range(pdf_reader.numPages):
    pageText = pdf_reader.getPage(page).extractText()
    matches = re.findall(pattern, pageText)
    if len(matches) > 0:
        phone_nr = matches[0]
        break

phone_nr
```
